```python
model = IsolationForest(contamination=0.05)  # Adjust contamination based on your needs
model.fit(features)

# Initialize a dictionary to store feature importance scores
feature_importance = {feature: 0 for feature in features.columns}

# Iterate through each tree in the model
for tree in model.estimators_:
    tree_depth = tree.tree_.max_depth

    # Traverse the tree and accumulate feature importance
    def traverse_tree(node, depth):
        if tree.tree_.children_left[node] != tree.tree_.children_right[node]:
            feature_index = tree.tree_.feature[node]
            feature_importance[features.columns[feature_index]] += 1
            traverse_tree(tree.tree_.children_left[node], depth + 1)
            traverse_tree(tree.tree_.children_right[node], depth + 1)

    traverse_tree(0, 0)  # Start traversing from the root node

# Normalize the importance scores
total_trees = len(model.estimators_)
for feature_name in feature_importance:
    feature_importance[feature_name] /= total_trees

# Sort features by importance in descending order
sorted_features = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)

# Print or visualize the most important features
print("Top features by importance:")
for feature_name, importance_score in sorted_features:
    print(f"{feature_name}: {importance_score:.4f}")
```
