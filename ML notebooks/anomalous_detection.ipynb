```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest
```

```python
label_encoder = LabelEncoder()
logon['user_encoded'] = label_encoder.fit_transform(logon['user'])
```

```python
logon['activity_encoded'] = label_encoder.fit_transform(logon['activity'])
logon['pc_encoded'] = label_encoder.fit_transform(logon['pc'])
```

```python
logon['day']=logon['date'].dt.strftime('%A')
```

```python
logon['day_encoded']= label_encoder.fit_transform(logon['day'])
```

```python
isolation_forest = IsolationForest(contamination=0.05)
```

```python
numerical_columns = ['month','year','hour','user_encoded', 'activity_encoded', 'pc_encoded','day_encoded']
```

```python
isolation_forest.fit(logon[numerical_columns])
```

```python
logon['anomaly_score'] = isolation_forest.decision_function(logon[numerical_columns])
logon['anomaly_label'] = isolation_forest.predict(logon[numerical_columns])
```
