```python
import pandas as pd
from dateutil.relativedelta import relativedelta
from pycaret.anomaly import *
from pycaret import *

# Load the dataset

dataset_path = "E:/Dataset/Computer Emergency Response Team (CERT) r6.2/r6.2/r6.2/New folder/combine/email_logon_device_http_user_merged_data.csv"
df = pd.read_csv(dataset_path)

# Initialize the PyCaret setup for anomaly detection
anomaly_setup = setup(df, session_id=1)

```

```python
# Create an LOF model
iforet_model = create_model('iforest')

# Assign anomaly scores to the dataset
df_with_anomaly_scores = assign_model(iforet_model)

# Append anomaly scores to the original table
df['anomaly_score'] = df_with_anomaly_scores['Anomaly_Score']

# Measure ROC for the LOF model
roc_results = plot_model(iforet_model)

# Display the ROC curve and AUC
print(roc_results)

# Display the resulting dataframe with anomaly scores
print(df.head())
```
